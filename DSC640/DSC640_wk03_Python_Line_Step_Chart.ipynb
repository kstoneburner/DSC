```python
import os
import sys
# //*** Imports and Load Data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#//*** Use the whole window in the IPYNB editor
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

#//*** Maximize columns and rows displayed by pandas
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
```

```python
raw_df = pd.read_excel("z_wk03_world-population.xlsm")
plt.style.available
```

```python


plot_df = raw_df

#plt.style.use('ggplot')

#plt.style.use('fivethirtyeight')
plt.style.use("Solarize_Light2")


size =7
scale=1.5
wedge = size
#colors = sns.color_palette('YlOrBr')[0:len(plot_df[x1])]

#//*** Plot properties
pp = {
    'df' : raw_df,
    'title' : "World Population: [1960-2010]",
    'xlabel' : "Population (Billions)", 
    'ylabel' : "Year", 
    'x_figsize' : 20,
    'y_figsize' : 10,
    'fontsize' : 20,
}

x_col = "Population"
y_col = "Year"


fig, ax = plt.subplots(ncols=2)
fig = plt.gcf()
fig.set_size_inches(pp['x_figsize'],pp['y_figsize'])
sns.lineplot(pp['df'][x_col]/1000000000,pp['df'][y_col], drawstyle='steps-pre',ax=ax[0])
#plt.xlabel(pp['xlabel'],fontsize=pp['fontsize']/1.5)

sns.lineplot(pp['df'][x_col]/1000000000,pp['df'][y_col], ax=ax[1])

plt.suptitle(pp['title'],fontsize=pp['fontsize']*1.5)
for i in [0,1]:
    ax[i].set_xlabel(pp['xlabel'],fontsize=pp['fontsize'])
    ax[i].set_ylabel(pp['ylabel'],fontsize=pp['fontsize'])
    ax[i].ticklabel_format(axis='both', style='plain', scilimits=None, useOffset=None, useLocale=None, useMathText=None)
    ax[i].tick_params(axis='both', labelsize=pp['fontsize']/1.5)

plt.text(1,-.1,'Python Graph', size=pp['fontsize']/1.5, transform=ax[0].transAxes)
plt.show()


```
