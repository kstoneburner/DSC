```python
import os
import sys
# //*** Imports and Load Data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#//*** Use the whole window in the IPYNB editor
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

#//*** Maximize columns and rows displayed by pandas
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
```

```python
raw_df = pd.read_excel("z_wk03_world-population.xlsm")
raw_df
```

```python


plot_df = raw_df

size =7
scale=1.5
wedge = size
#colors = sns.color_palette('YlOrBr')[0:len(plot_df[x1])]

#//*** Plot properties
pp = {
    'df' : raw_df,
    'title' : "World Population: [1960-2010]",
    'xlabel' : "Population (Billions)", 
    'x_figsize' : 10,
    'y_figsize' : 10,
    'fontsize' : 20,
}

x_col = "Population"
y_col = "Year"


fig, ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(pp['x_figsize'],pp['y_figsize'])

ax.plot(pp['df'][x_col]/1000000000,pp['df'][y_col])
ax.ticklabel_format(axis='both', style='plain', scilimits=None, useOffset=None, useLocale=None, useMathText=None)
plt.title(pp['title'],fontsize=pp['fontsize'])
plt.xlabel(pp['xlabel'],fontsize=pp['fontsize']/1.5)
plt.show()

plot_df = raw_df

size =7
scale=1.5
wedge = size
x_col = "Population"
y_col = "Year"

#colors = sns.color_palette('YlOrBr')[0:len(plot_df[x1])]
fig, ax = plt.subplots()

fig = plt.gcf()
fig.set_size_inches(pp['x_figsize'],pp['y_figsize'])

ax.step(pp['df'][x_col]/1000000000,pp['df'][y_col])
ax.ticklabel_format(axis='both', style='plain', scilimits=None, useOffset=None, useLocale=None, useMathText=None)
plt.title(pp['title'],fontsize=pp['fontsize'])
plt.xlabel(pp['xlabel'],fontsize=pp['fontsize']/1.5)
plt.show()
```
